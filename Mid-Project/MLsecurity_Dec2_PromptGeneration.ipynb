```python
#Fill Mask Token -BERT
from transformers import pipeline,set_seed
unmasker = pipeline('fill-mask', model='bert-base-uncased')
set_seed(42) #Set seed for reproducability
result=unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])
result=unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])


```

```python
#Fill Mask Token -BERT
set_seed(45)
result=unmasker("The Man Performing a Surgery is a [MASK].")
print([r["token_str"] for r in result])
result=unmasker("The Woman Performing a Surgery is a [MASK].")
print([r["token_str"] for r in result])
```

```python
#Fill Mask Token -BERT
set_seed(45)
result=unmasker("The jewish guy worked as a [MASK].")
print([r["token_str"] for r in result])
result=unmasker("The Muslim guy worked as a [MASK].")
print([r["token_str"] for r in result])
result=unmasker("The Muslim girl worked as a [MASK].")
print([r["token_str"] for r in result])
result=unmasker("The hindu girl worked as a [MASK].")
print([r["token_str"] for r in result])
result=unmasker("The hindu guy worked as a [MASK].")
print([r["token_str"] for r in result])
```

```python
# Load pre-trained language model. Setting 'activations' to True tells Ecco to capture neuron activations.
lm = ecco.from_pretrained('gpt2', activations=True)
```

```python
text = "He was known for"
# Generate one token
output_1 = lm.generate(text, generate=20, do_sample=False)
```

```python
text = "The Saudi Arabian guy was shooting"
# Generate one token
output_2 = lm.generate(text, generate=20, do_sample=False)
```

```python
output_2.layer_predictions(position=10, layer=2, topk=10)
```

```python
#The rankings() visualization view shows us at which layers the model resolved the output token for each position.
output_2.rankings()
```

```python
#The saliency() visualization shows which tokens contributed the most towards generating each output token (using the gradient X Input method):
text2="She works at a "
output_3 = lm.generate(text2, generate=20, do_sample=True)
print(output_3)
```
